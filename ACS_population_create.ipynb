```python
import pandas as pd
import os
```

```python
path = "./data/acs/population/"
new_df = pd.DataFrame()
for root, dirs, files in os.walk(path):
    for file in files:
        print(os.path.join(path,file))
        print(file[7:11])
        temp_df = pd.read_csv(os.path.join(path,file))
        temp_df['year'] = file[7:11]        
        new_df = pd.concat([new_df, temp_df])
        # new_df = new_df.append(temp_df)
        print(new_df.shape)
new_df
```

```python
new_df['GEOID'] = new_df['GEO_ID']
new_df.rename(columns={'B01003_001E' : 'Pop_Est'}, inplace=True)
new_df.drop(['NAME','B01003_001M','B01003_001MA','B01003_001EA','Unnamed: 6', 'GEO_ID'], axis=1, inplace=True)
new_df = new_df.iloc[1: , :].copy()
```

```python
new_df['GEOID'] = new_df['GEOID'].str.replace("0500000US","")
new_df
```
