# R1, R2, R3 Universities

Three tables in below wikipedia link (R1,R2, R3) with the following features.
- Institution
- Control
- City
- State

```python
url = 'https://en.wikipedia.org/wiki/List_of_research_universities_in_the_United_States'
```

```python
r = requests.get(url, headers = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'})
```

```python
r
```

```python
myhtml = BeautifulSoup(r.text, 'html.parser')
```

```python
r1table = myhtml.find('table',{'class':"wikitable"})
```

```python
r1=pd.read_html(str(r1table))
# convert list to dataframe
r1=pd.DataFrame(r1[0])
r1['level'] = 'r1'
r1
```

```python
tables = myhtml.find_all('table',{'class':"wikitable"})
all=pd.read_html(str(tables))
# convert list to dataframe
r2=pd.DataFrame(all[1])
r2['level'] = 'r2'
r2
```

```python
r3=pd.DataFrame(all[2])
r3['level'] = 'r3'
r3
```

```python
universities = pd.concat([r1, r2, r3])
```

```python
save_location = 'C:\\Users\\trent\\Documents\\Capstone\\Data\\Research Universities'
```

```python
os.chdir(save_location)
```

```python
universities.to_csv('universities.csv')
```
