## Preparing grant and pregant for concatenation 

```python
import pandas as pd
import numpy as np
```

```python
granted = pd.read_csv("granted_patents.csv", dtype = {'GEOID' : 'str'}, low_memory = False) #, index_col=[0])
granted
```

```python
granted['already_granted']= 1
granted
```

```python
pre_grant = pd.read_csv("pre_grant_stats.csv", index_col=[0], dtype = {'GEOID': 'str'}, low_memory = False)
pre_grant
```

```python
pre_grant['already_granted']= 0
pre_grant
```

```python
#filter for just the years we want 2009 - 2019 
pre_grant = pre_grant.query("year_x >=2009 & year_x <= 2019").sort_values(by = 'year_x')
pre_grant
```

```python
#pre_grant.type_x.unique()
```

```python
#organization will be renames assignee -> despite not being granted 
pre_grant = pre_grant.rename({'organization':'assignee', 'application_number_x':'application_number',
                  'year_x': 'application_year', 'male_inventors': 'men_inventors',
                  'female_inventors': 'women_inventors'},
                 axis =1)
                  
                  
pre_grant
```

```python
pre_grant.drop('type_x', axis = 1, inplace = True)
pre_grant
```

```python
pre_grant.columns
```

```python
pre_grant = pre_grant.assign(patent_number = np.nan, grant_year = np.nan, inventors = pre_grant.team_size, ipc_section = np.nan)
pre_grant = pre_grant[['patent_number', 'assignee', 'grant_year', 'application_year',
       'application_number','GEOID', 'ipc_section', 'team_size', 'inventors', 'men_inventors',
       'women_inventors', 'already_granted']]
```

```python
granted = granted[['patent_number', 'assignee', 'grant_year', 'application_year',
       'application_number', 'GEOID', 'ipc_section', 'team_size', 'inventors', 'men_inventors',
       'women_inventors', 'already_granted']]
```

# Concat grant and pregrant

```python
all_patents = pd.concat([granted, pre_grant], ignore_index=True)
sum(all_patents.duplicated())
```

```python
all_patents = all_patents[~all_patents.duplicated()]
```
