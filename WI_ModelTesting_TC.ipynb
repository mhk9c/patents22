```python
def true_pos(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1

    return(TP, FP, TN, FN)
```

```python
def log_reg(features, x_train, y_train, x_test, y_test, polynomials = None, degree = 1):

    if polynomials == None:
        x_test = x_test[features]
        model = LogisticRegression(solver='saga')
        logreg = model.fit(x_train[features], y_train)
        yhat = model.predict_proba(x_test)[:,1]
        y_hat_pred = model.predict(x_test)
        y_hat_pred_3 = (model.predict_proba(x_test)[:,1] >= 0.3).astype(bool)
        y_hat_pred_7 = (model.predict_proba(x_test)[:,1] >= 0.7).astype(bool)
        y_hat_pred_9 = (model.predict_proba(x_test)[:,1] >= 0.9).astype(bool)
        auc = metrics.roc_auc_score(y_test, yhat).round(3)
        f1_5 = metrics.f1_score(y_test, y_hat_pred).round(3)
        tp_5 = metrics.recall_score(y_test, y_hat_pred).round(3)
        f1_3 = metrics.f1_score(y_test, y_hat_pred_3).round(3)
        tp_3 = metrics.recall_score(y_test, y_hat_pred_3).round(3)
        f1_7 = metrics.f1_score(y_test, y_hat_pred_7).round(3)
        tp_7 = metrics.recall_score(y_test, y_hat_pred_7).round(3)
        f1_9 = metrics.f1_score(y_test, y_hat_pred_9).round(3)
        tp_9 = metrics.recall_score(y_test, y_hat_pred_9).round(3)
        lst_dict.append({'auc':auc, 'f1_.5': f1_5, 'tp_.5': tp_5,
                         'f1_.3': f1_3, 'tp_.3': tp_3,
                         'f1_.7': f1_7, 'tp_.7': tp_7,
                         'f1_.9': f1_9, 'tp_.9': tp_9,
                         'features':features, 'degree': degree, 
                         'polynomial_features':polynomials})
        return lst_dict
    
    else:
        x_test = x_test[features]
        x_train = x_train[features]
        poly = DataFrameMapper([(polynomials, 
                               PolynomialFeatures(degree))
                               ])
        x_train = poly.fit_transform(x_train)
        x_test = poly.fit_transform(x_test)
        model = LogisticRegression(solver='saga')
        logreg = model.fit(x_train, y_train)
        yhat = model.predict_proba(x_test)[:,1]
        y_hat_pred = model.predict(x_test)
        y_hat_pred_3 = (model.predict_proba(x_test)[:,1] >= 0.3).astype(bool)
        y_hat_pred_7 = (model.predict_proba(x_test)[:,1] >= 0.7).astype(bool)
        y_hat_pred_9 = (model.predict_proba(x_test)[:,1] >= 0.9).astype(bool)
        auc = metrics.roc_auc_score(y_test, yhat).round(3)
        f1_5 = metrics.f1_score(y_test, y_hat_pred).round(3)
        tp_5 = metrics.recall_score(y_test, y_hat_pred).round(3)
        f1_3 = metrics.f1_score(y_test, y_hat_pred_3).round(3)
        tp_3 = metrics.recall_score(y_test, y_hat_pred_3).round(3)
        f1_7 = metrics.f1_score(y_test, y_hat_pred_7).round(3)
        tp_7 = metrics.recall_score(y_test, y_hat_pred_7).round(3)
        f1_9 = metrics.f1_score(y_test, y_hat_pred_9).round(3)
        tp_9 = metrics.recall_score(y_test, y_hat_pred_9).round(3)
        lst_dict.append({'auc':auc, 'f1_.5': f1_5, 'tp_.5': tp_5,
                         'f1_.3': f1_3, 'tp_.3': tp_3,
                         'f1_.7': f1_7, 'tp_.7': tp_7,
                         'f1_.9': f1_9, 'tp_.9': tp_9,
                         'features':features, 'degree': degree, 
                         'polynomial_features':polynomials})
        return lst_dict
```
